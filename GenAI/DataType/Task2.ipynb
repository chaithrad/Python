```python
categories = ["Electronics", "Electronics","Electronics","Stationery","Acessories","Electronics","Stationery","Electronics"]
# 1. create set a set of catergories
catergory_set = set(categories)
print("Set of categories:", catergory_set)
# 2. Add a new category to the set
catergory_set.add("Clothing")
print("Set of categories after adding 'Clothing':", catergory_set)
# TRy adding a duplicate category to the set
catergory_set.add("Electronics")
print("Set of categories after adding 'Electronics':", catergory_set)
# 3. Check whether a category exists in the set
category_to_check = "Stationery"
if category_to_check in catergory_set:
    print(f"'{category_to_check}' exists in the set.")
else:
    print(f"'{category_to_check}' does not exist in the set.")  
# 4. extra : count unique categories in the original list
unique_category_count = len(catergory_set)
print("Number of unique categories in the original list:", unique_category_count)

```
