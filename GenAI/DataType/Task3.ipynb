```python
#1. create dictionary
price_dict = { "laptop": 800, "phone": 500, "tablet": 300,"notebook": 1000, "monitor": 200 , "keyboard": 50 }
print(price_dict)
# 2. add a new product to the dictionary
price_dict["headphones"] = 150
print(price_dict)
# 3.update an existing product price in the dictionary
price_dict["laptop"] = 750
print(price_dict)
# 4. remove a product safely from the dictionary
product_to_remove = "tablet"
removed_price = price_dict.pop(product_to_remove, None)
if removed_price is not None:
    print(f"Removed {product_to_remove} with price {removed_price}")    
else:
    print(f"{product_to_remove} not found in the dictionary.")
print(price_dict)
#5. Avergae price of all products in the dictionary
average_price = sum(price_dict.values()) / len(price_dict)
print(f"Average price of all products: {average_price}")
# 5. Extra : max and min price of all products in the dictionary
max_price_product = max(price_dict, key=price_dict.get)
min_price_product = min(price_dict, key=price_dict.get)
print(f"Product with max price: {max_price_product} - Price: {price_dict[max_price_product]}")
print(f"Product with min price: {min_price_product} - Price: {price_dict[min_price_product]}")
    
```
