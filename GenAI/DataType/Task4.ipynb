```python
# product list
products = ["laptop", "phone", "tablet", "notebook", "monitor", "water bottle"]
#  CAtergories list
categories = ["electronics", "electronics", "electronics", "stationery", "electronics", "accessories"]
prices = [1000, 5000, 3000, 5000, 2000, 100]
# 1. create catalog
catalog = []
for i in range(len(products)):
    catalog.append((products[i], categories[i]))
print(catalog)
# 2. create category_to_produts dictionary
category_to_products = {}
for product, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(product)
print(category_to_products)
# 3.categoru with max number of prodtucts
max_category = ""
max_count = 0
for category, products in category_to_products.items():
    if len(products) > max_count:
        max_count = len(products)
        max_category = category
print(f"Category with max products: {max_category} - Count: {max_count}")
```
